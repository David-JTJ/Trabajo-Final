```python
gen = []
L = []
for i in lista_pag:
    datos = df_sheet_map[i]
    L = datos.iloc[1:,4].to_list()    
    gen.extend(L)
print(gen)
```

```python
gen2 = list(set(pd.Series(gen).str.upper()))
del gen2[0]
print(len(gen2))
```
